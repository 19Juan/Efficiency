# Eficiencia

```python
import pandas as pd
df = pd.read_csv('../data/processed/BD.csv', sep=';', encoding='latin1')
df.info()
df.shape
#No se cuenta con valores nulos.
```

```python
#AHora quiero pasar todas las variables, menos marca, a numericas
df['AGUA EN CRUDOS'] = pd.to_numeric(df['AGUA EN CRUDOS'], errors='coerce')
df['RELACION DE AGUA CRUDOS'] = pd.to_numeric(df['RELACION DE AGUA CRUDOS'], errors='coerce')
df['RELACION DE AGUA MEZCLAS'] = pd.to_numeric(df['RELACION DE AGUA MEZCLAS'], errors='coerce')
df['VOLUMEN TOTAL DE FILTRACION'] = pd.to_numeric(df['VOLUMEN TOTAL DE FILTRACION'], errors='coerce')
df['DIFERENCIAL T vapor y mosto'] = pd.to_numeric(df['DIFERENCIAL T vapor y mosto'], errors='coerce')
df['Extracto inicio coccion'] = pd.to_numeric(df['Extracto inicio coccion'], errors='coerce')
df['Extracto  fin coccion'] = pd.to_numeric(df['Extracto  fin coccion'], errors='coerce')
df[' Rata Evap '] = pd.to_numeric(df[' Rata Evap '], errors='coerce')
df['Tiempo total estres termico'] = pd.to_numeric(df['Tiempo total estres termico'], errors='coerce')
df['Volumen mosto frio'] = pd.to_numeric(df['Volumen mosto frio'], errors='coerce')
df['EO mosto frio'] = pd.to_numeric(df['EO mosto frio'], errors='coerce')
df['pH mosto frio'] = pd.to_numeric(df['pH mosto frio'], errors='coerce')
df['Kg Extracto entrada'] = pd.to_numeric(df['Kg Extracto entrada'], errors='coerce')
df['Eficiencia'] = pd.to_numeric(df['Eficiencia'], errors='coerce')
```

```python
df.info()
```

```python
#Ahora a borrar filas con nulos:
df = df.dropna()
df
```

```python
df.info()
```

```python
#Anlisis univariante
#variable categrica
import matplotlib.pyplot as plt
import seaborn as sns

fig, axis = plt.subplots(1, figsize=(15, 5))

sns.histplot(df['MARCA'], ax=axis).set(ylabel='Frecuencia')
plt.title('Distribucin de la variable MARCA')

plt.grid(axis='y')

plt.show()
```

```python
df['MARCA'] = pd.factorize(df['MARCA'])[0]

sns.heatmap(df[['MARCA','Kg Extracto entrada',
'RELACION DE AGUA CRUDOS',
'RELACION DE AGUA MEZCLAS',
'VOLUMEN TOTAL DE FILTRACION',
'DIFERENCIAL T vapor y mosto',
'Extracto inicio coccion',
'Extracto  fin coccion',
'EO mosto frio',
'Eficiencia']].corr(), annot=True,fmt='.1f')
plt.title('Correlacin entre variables')
plt.tight_layout()
plt.show()

#Ahora quiero guardar la imagen del heat map
plt.savefig('../data/processed/heatmap.png', dpi=300, bbox_inches='tight')


```
